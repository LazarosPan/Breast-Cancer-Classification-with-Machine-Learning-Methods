```python
# from functools import reduce
# data_frames = [allf_default, sf_default, pcaf_default]

# # if you want to fill the values that don't exist in the lines of merged dataframe simply fill with required strings as

# # df_merged = reduce(lambda  left,right: pd.merge(left,right,on=['algorithm'],
# #                                             how='outer'), data_frames) #.fillna('void')
# df_merged

#or
# use concat to combine more than two DataFrames
f1_default = pd.concat([allf_default[['algorithm','f1 score']].set_index('algorithm'), sf_default[['algorithm','f1 score']].set_index('algorithm'), pcaf_default[['algorithm','f1 score']].set_index('algorithm')], axis=1)

time_default = pd.concat([allf_default[['algorithm','Execution Time (seconds)']].set_index('algorithm'), sf_default[['algorithm','Execution Time (seconds)']].set_index('algorithm'), pcaf_default[['algorithm','Execution Time (seconds)']].set_index('algorithm')], axis=1)

f1_div_time_default = pd.concat([allf_default[['algorithm','f1/time']].set_index('algorithm'), sf_default[['algorithm','f1/time']].set_index('algorithm'), pcaf_default[['algorithm','f1/time']].set_index('algorithm')], axis=1)
```

```python
# colors = ['#5cb85c', '#5bc0de', '#d9534f']
sns.set(style="darkgrid",font_scale = 1.6)
mylabels=['All Features (30)','Selected Features (7)','Extracted Features (10)']
_ = f1_default.plot(kind='bar', figsize=(30, 10), rot=0, ylabel='F1 Score', title="F1 Score of the default algorithms applied to the feature selection and extraction techniques")
_.set_ylim(0.9,0.999999999)
_.legend(labels=mylabels);
for i, rect in enumerate(_.containers):
    for j, bar in enumerate(rect):
        _.text(bar.get_x() + bar.get_width() / 2,
                bar.get_height() + 0.001,
                f"{bar.get_height():.3f}",
                ha='center',
                va='bottom',
                rotation=90)


_ = time_default.plot(kind='bar', figsize=(30, 10), rot=0, ylabel='Execution Time (seconds)', title="Execution time of the default algorithms applied to the feature selection and extraction techniques")
_.set_ylim(0,5.3)
_.legend(labels=mylabels);
for i, rect in enumerate(_.containers):
    for j, bar in enumerate(rect):
        _.text(bar.get_x() + bar.get_width() / 2,
                bar.get_height() + 0.001,
                f"{bar.get_height():.3f}",
                ha='center',
                va='bottom',
                rotation=90)

_ = f1_div_time_default.plot(kind='bar', figsize=(30, 10), rot=0, ylabel='F1 score / Execution Time (seconds)', title="Efficiency of the default algorithms applied to the feature selection and extraction techniques (Higher is better)")
_.set_ylim(0,17)
_.legend(labels=mylabels);
for i, rect in enumerate(_.containers):
    for j, bar in enumerate(rect):
        _.text(bar.get_x() + bar.get_width() / 2,
                bar.get_height() + 0.001,
                f"{bar.get_height():.3f}",
                ha='center',
                va='bottom',
                rotation=90)

# SOURCE : https://stackoverflow.com/questions/63493530/how-to-plot-and-annotate-a-grouped-bar-chart
```

* Tuned Algorithms

```python
allf_tuned['f1/time'] = allf_tuned['f1 score']/allf_tuned['Execution Time (seconds)']
sf_tuned['f1/time'] = sf_tuned['f1 score']/sf_tuned['Execution Time (seconds)']
pcaf_tuned['f1/time'] = pcaf_tuned['f1 score']/pcaf_tuned['Execution Time (seconds)']
```

```python
f1_tuned = pd.concat([allf_tuned[['algorithm','f1 score']].set_index('algorithm'), sf_tuned[['algorithm','f1 score']].set_index('algorithm'), pcaf_tuned[['algorithm','f1 score']].set_index('algorithm')], axis=1)

time_tuned = pd.concat([allf_tuned[['algorithm','Execution Time (seconds)']].set_index('algorithm'), sf_tuned[['algorithm','Execution Time (seconds)']].set_index('algorithm'), pcaf_tuned[['algorithm','Execution Time (seconds)']].set_index('algorithm')], axis=1)

f1_div_time_tuned = pd.concat([allf_tuned[['algorithm','f1/time']].set_index('algorithm'), sf_tuned[['algorithm','f1/time']].set_index('algorithm'), pcaf_tuned[['algorithm','f1/time']].set_index('algorithm')], axis=1)
```

```python
_ = f1_tuned.plot(kind='bar', figsize=(30, 10), rot=0, ylabel='F1 Score', title="F1 Score of the tuned algorithms applied to the feature selection and extraction techniques")
_.set_ylim(0.9,0.9999999)
_.legend(labels=mylabels);
for i, rect in enumerate(_.containers):
    for j, bar in enumerate(rect):
        _.text(bar.get_x() + bar.get_width() / 2,
                bar.get_height() + 0.001,
                f"{bar.get_height():.3f}",
                ha='center',
                va='bottom',
                rotation=90)

_ = time_tuned.plot(kind='bar', figsize=(30, 10), rot=0, ylabel='Execution Time (seconds)', title="Execution time of the tuned algorithms applied to the feature selection and extraction techniques")
_.set_ylim(0,420)
_.legend(labels=mylabels);
# _.ticklabel_format(style='sci', axis='y', scilimits=(0,0))
for i, rect in enumerate(_.containers):
    for j, bar in enumerate(rect):
        _.text(bar.get_x() + bar.get_width() / 2,
                bar.get_height() + 0.001,
                f"{bar.get_height():.3f}",
                ha='center',
                va='bottom',
                rotation=90)

_ = f1_div_time_tuned.plot(kind='bar', figsize=(30, 10), rot=0, ylabel='F1 score / Execution Time (seconds)', title="Efficiency of the tuned algorithms applied to the feature selection and extraction techniques (Higher is better)")
_.set_ylim(0,0.35)
_.legend(labels=mylabels);
for i, rect in enumerate(_.containers):
    for j, bar in enumerate(rect):
        _.text(bar.get_x() + bar.get_width() / 2,
                bar.get_height() + 0.001,
                f"{bar.get_height():.3f}",
                ha='center',
                va='bottom',
                rotation=90)
```
