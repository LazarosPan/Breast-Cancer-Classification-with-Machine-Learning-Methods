## Explanation of the figures

The results of the algorithms on the **default parameters** show that:

* F1 scores for all algorithms range from 0.915 to 0.981 with computation times ranging from 0.062 to 4.628 seconds. 

* The linear classifiers Ridge, LDA and GNB keep their performance stable with the reduced features as they are very close in F1 score and execution time. It is observed that Ridge and LDA achieved the same scores with 7 and 10 features while being quite close with all features.

* The F1 score of 4 of the 5 algorithms related to decision trees (LGBM, Random Forest, XGBoost and Decision Tree) seem to be reduced when trained with the 7 features. This is logically because a lot of information was lost by reducing the features, whereas with the features from PCA LGBM and Decision Tree had improved results.

* The AdaBoost classifier, which takes as its base estimator the tree classifiers trained on each feature set, seems to achieve better results with the weak decision trees as described in the theory.

* The 7 selected features and the 10 features from PCA generally lead to faster computation times.

* The use of the 10 features from PCA tends to lead to better performance in terms of F1 score, with the best performance being achieved by the SVC (0.981 in 0.093 sec) and MLP (0.981 in 3.7 sec) algorithms.


The results of the algorithms on **configured parameters** show that:

* F1 scores for all algorithms range from 0.927 to 0.983 with computation times ranging from 3.455 to 352.877 seconds. 

* The linear Ridge and LDA classifiers hold their performance constant with the reduced features, as they are very close in F1 score and runtime. GNB seems to have fared better with the features than PCA. It is also observed that Ridge and LDA have exactly the same scores for all feature sets, but LDA is significantly faster.

* The F1 scores of the algorithms related to decision trees (LGBM, Random Forest, XGBoost and Decision Tree) this time seem to have improved with the 7 features while the opposite happened with the features from PCA.

* The AdaBoost classifier, again seems to have done better with the weak decision tree. Although in the decision trees the score was the same in training with all features and those from PCA, the one with all features did much better, which can be attributed to the random parameter search. 

* The 7 selected features and the 10 features from PCA generally lead to faster computation times.

* The use of the 10 features from PCA tends to lead to better performance in terms of F1 score, with the best performance being achieved by the SVC (0.981 in 9.945 sec) and MLP (0.981 and 0.983 with testing) algorithms. The improvement of the SGD (0.979 in 4.961 sec) and KNN (0.979 in 87.843) algorithms is also noteworthy.

In order to show the improvement (or not) of the algorithms with the tuned parameters compared to those with the default parameters, the last figure was created. This was done by calculating the **difference in F1 scores and runtimes between the two groups of algorithms**. The following conclusions are drawn from this figure:

* Not all algorithms improved, instead some of them dropped their performance. In particular, AdaBoost dropped its performance on all feature sets. Random Forests and GNB dropped their performance on 2 of the 3 feature sets.

* Here we can clearly see how much the increase in Decision Tree's performance affects the decrease in AdaBoost's score.

* Computation times increased tremendously, especially in some cases such as in AdaBoost, Decision Trees, KNN, Random Forests and XGBoost.

* The KNN, MLP and SGD algorithms increased their performance on all feature sets. Decision Tree, QDA and XGBoost improved on 2 of the 3 feature sets.

* Great improvement was achieved by the SGD algorithm with all features (+0.033 to +4.879 sec), but also by the Decision Trees on the 7 selected features (+0.019 to +42.525 sec).

* The score of LDA (+0.002) and Ridge (-0.003) was minimally changed when trained with all features, while the SVC algorithm was not changed at all.

## Conclusion

It is observed that most algorithms did better on the F1 score with the features extracted from PCA. This is because just 10 features out of 30 contain about 95.33% of the total information and thus the algorithms learned the classification "patterns" more easily. 

Also, significant results, mainly in terms of runtime, were also achieved by the algorithms with the 7 selected features as in most cases they were faster than the other 2 feature sets.

In general, the best performance seems to have been achieved by the nonlinear classifiers, namely MLP, SVC, KNN, SGD and QDA. In contrast, the linear classifiers, namely Ridge, LDA and GNB, performed poorly , but had a fairly short runtime. The algorithms related to decision trees, namely LGBM, XGboost, AdaBoost, Random Forest and Decision Tree, had relatively moderate results but very long runtimes.

In summary, the highest F1 scores were achieved by the following algorithms on the features from PCA:

1. **MLP**
    * 0.983 and optimization with trial and error.
    * 0.981 in 3.7 seconds with default parameters.
    

2. **SVC**
    * 0.981 in 0.093 seconds with default parameters.
    * 0.981 in 9.945 seconds with tuned parameters.

In conclusion, the results of the experiments showed that MLP had the highest F1 score, making it the best performing algorithm in terms of accuracy. However, this came at the cost of more computational time due to the optimization with trials. On the other hand, SVC had a slightly lower F1 score, but was much more efficient with a computational time of only 0.093 seconds.

Nevertheless, when it comes to real-world applications in the context of breast cancer diagnosis, accuracy is paramount. Thus, MLP is still the preferred choice, as every small improvement in the F1 score can potentially save lives. It is a trade-off between computational efficiency and accuracy, but in this scenario, a higher F1 score is worth the extra time spent on tuning.
